```python
import os
import cv2
import numpy as np 


# Example of the file system path
image_path = r"E:\Ragib\Noman_vai\tea_leaves\dataset\SUST\original_dataset_1\Tea Images\Healthy"
image_new_path = r"E:\Ragib\Noman_vai\tea_leaves\dataset\SUST\filtered\non_local_means_denoising\Healthy"

def list_images():
    if os.path.exists(image_path):
        print("Path exists. Listing files:")
        image_list = os.listdir(image_path)
        new_image_list = os.listdir(image_new_path)
        for file in image_list:
            image = cv2.imread(image_path + '/' + file)
            if "nlmd_"+file in new_image_list:
                print("already filtered "+ file)
                continue
            else:
                print(file)
                denoised_image = cv2.fastNlMeansDenoisingColored(image, None, 10, 10, 7, 21) 
                cv2.imwrite(r"E:\Ragib\Noman_vai\tea_leaves\dataset\SUST\filtered\non_local_means_denoising\Healthy/nlmd_" + file  , denoised_image  )
        return image_list
    else:
        print(f"Error: Path '{image_path}' does not exist.")
        return []

# Call the function and assign the result to image_list
image_list = list_images()
print("Health completed.")
```
